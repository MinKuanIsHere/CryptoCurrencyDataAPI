# Bitfinex
https://docs.bitfinex.com/reference/rest-public-candles

```python
import requests
import json
import datetime
import pytz
import pandas as pd

def Str2Ms(date_str):
    dt_obj = datetime.datetime.fromisoformat(date_str) # Convert string to datetime object in UTC
    dt_obj_utc = dt_obj.replace(tzinfo=pytz.UTC)
    timestamp_seconds = dt_obj_utc.timestamp() # Convert datetime object to timestamp (in seconds)
    timestamp_milliseconds = int(timestamp_seconds * 1000) # Convert seconds to milliseconds
    return timestamp_milliseconds

def BitFinex(sym, freq, time_start, time_end):
    time_start = Str2Ms(time_start) # convert date str to MS int
    time_end = Str2Ms(time_end) # convert date str to MS int
    url = f"https://api-pub.bitfinex.com/v2/candles/trade:{freq}:t{sym}USD/hist"
    headers = {"accept": "application/json"}
    params = {'sort':1, 'start': time_start, 'end': time_end, 'limit':2000}
    response = requests.get(url, headers=headers, params = params)
    print(response)
    data = json.loads(response.text)
    df = pd.DataFrame(data, columns=['time', 'open', 'high', 'low', 'close', 'volume'])
    df["time"] = pd.to_datetime(df['time'], unit='ms')
    return df

sym = "BTC"
freq = "1m"
time_start = "2023-03-01T00:00:00"
time_end = "2023-03-01T23:59:59"
df = BitFinex(sym, freq, time_start, time_end)
df
```
